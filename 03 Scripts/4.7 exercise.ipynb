```python
#Define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

## If-Statements with For-Loops

```python
#On which day most orders take place--- frequency of orders on days of week

df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# 0 - Saturday is the busiest day (the day when the most orders take place)
# 4 - Wednesday is the least busy day
```

```python
# Create a new column, busiest day, that will contain one of three different values: Busiest day,Least busy,and â€œRegularly busy.
# How? 
# Creating a for-loop that will run through every row in the â€œorders_day_of_week column, 
# compare values with the busiest and slowest days, and assign it the corresponding string value.
```

```python
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# This for loop will check every single value under the 'orders_day_of_week'.
# Then, it will apply the conditions above. After that, it will push the condition to the 'result = []' list 
```

```python
result
```

```python
df_ords_prods_merge['busiest_day'] = result
```

```python
#Question 2. This label should correspond with the two busiest days of the week as opposed to the single busiest day. 
# At the same time, they also like to know the two slowest days. Create a new column for this using a suitable method.
```

```python
results = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value in (0, 1):
        results.append("Busiest days")
    elif value in(3, 4):
        results.append("Slowest days")
    else:
        results.append("Regular days")
```

### Explanation
 In this task, we’re categorizing the days of the week based on customer activity (order frequency). The days are represented as numbers in the orders_day_of_week column, where:
From the analysis, we already know:
Saturday (0) and Sunday (1) are the busiest days
Wednesday (3) and Thursday (4) are the slowest days
The rest are considered regular days

We used a for loop to go through each value in the orders_day_of_week column and classify it accordingly
 How It Works:
if value in (0, 1): This checks if the day is Saturday or Sunday. If so, we label it "Busiest days".
elif value in (3, 4): If it’s Wednesday or Thursday, we label it "Slowest days".
else: For any other day (Monday, Tuesday, Friday), we label it "Regular days".

The expression value in (0, 1) is a shorthand way of saying value == 0 or value == 1. This is where the logical operator or is being used — to check for multiple possible conditions at once.

```python
#Creating a new column 
df_ords_prods_merge["busiest_days_summary"] = results
```

```python
#Question 4. The senior technical officer at Instacart wants you to identify the busiest hours of the day. 
#Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” 
#Create a new column containing these labels called “busiest_period_of_day.”
```

```python
#checking order frequency by hour 
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
#Categorize hours- split the 24 hours (0–23) based on order frequency into three groups:
#Most orders:These hours had the highest number of orders: 10, 11, 14, 15, 13, 12, 16, 9
#Fewest orders:These had the lowest order volumes: 23, 6, 0, 1, 5, 2, 4, 3
#Average orders:All the remaining hours fall into this category: 7, 8, 17, 18, 19, 20, 21, 22
```

```python
#use loop to assign label for each hour
result_hour = []

for value in df_ords_prods_merge['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_hour.append("Most orders")
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        result_hour.append("Fewest orders")
    else:
        result_hour.append("Average orders")
```

```python
#Creating the column 'busiest_period_of_day' and assigning to it the values stored in 'result_hour'
df_ords_prods_merge['busiest_period_of_day'] = result_hour
```

```python
#Question-5 Print the frequency for this new column
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
