## Merging the dataframes

#### Observation
##### The dataframe already had a _merge column, so it had to be renamed as indicator = merge_status

```python
#Conducting a test merge without overwriting
pd.merge(df_ord_prod_combined,df_prods, on = ['product_id'], indicator = 'merge_status')
```

##### Dropping the _merge column as it is not necessary any more

```python
#drop _merge column
df_ord_prod_combined.drop('_merge', axis = 1, inplace = True)
```

```python
#Merging the dataframes df_ord_prod_combined and df_prods
ords_prods_merge = df_ord_prod_combined.merge(df_prods, on = ['product_id'], indicator = 'merge_status')
```
