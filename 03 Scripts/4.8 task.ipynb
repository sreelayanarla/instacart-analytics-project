## Create a loyalty flag 
### question 4

```python
#step 1 -- Use transform() to create a new column that shows the maximum number of orders (order_number) each customer has made.

ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform('max')
```

```python
#chcek output
ords_prods_merge.head()
```

```python
#Step 2 -- .Use loc() to create another column that flags these customers as "loyal" based on a condition (e.g., more than 40 orders).
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Examining Value Counts on 'loyalty_flag'
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#Checking the Updated Output
ords_prods_merge.head(15)
```

### difference between the spending habits of the three types of customers
### question 5

```python
# Determining Basic Statistics of Spending Habits by Loyalty Category
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

##### All three groups have the same minimum and maximum prices, suggesting they have access to the same range of products.

The mean price is slightly lower for Loyal customers (7.77) compared to New (7.80) and Regular (7.80) customers.

This small difference could indicate that loyal customers are more price-conscious, possibly due to repeated purchases or familiarity with cheaper options.

While the difference is subtle, it does suggest that loyal customers may tend to spend slightly less per product on average than new or regular customers.

### Create a spending flag
### question 6

```python
# Creating column for average spender with transform function
ords_prods_merge['average_spend']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Checking Relevant Columns 
ords_prods_merge[['user_id', 'average_spend']].head()
```

```python
# Creating flag for low and high spender
ords_prods_merge.loc[ords_prods_merge['average_spend']<10,'spender_flag']='Low_spender'

ords_prods_merge.loc[ords_prods_merge['average_spend']>= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check spender flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
pd.options.display.max_rows = 100
```

```python
ords_prods_merge[['user_id', 'average_spend', 'spender_flag', 'prices']].head(100)
```

### Create an Order Frequency Flag
### question 7

```python
#Create a column for frequent users
ords_prods_merge['Customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking Relevant Columns 
ords_prods_merge[['user_id', 'days_since_prior_order', 'Customer_frequency']].head(15)
```

```python
# Creating flag on order frequency
ords_prods_merge.loc[ords_prods_merge['Customer_frequency'] > 20, 'frequency_flag'] = 'Non_frequent_customer'

ords_prods_merge.loc[(ords_prods_merge['Customer_frequency']>10) & (ords_prods_merge['Customer_frequency']<=20),'frequency_flag'] = 'Regular customer'

ords_prods_merge.loc[ords_prods_merge['Customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Check frequency flags

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id','days_since_prior_order', 'Customer_frequency', 'frequency_flag']].head(100)
```
